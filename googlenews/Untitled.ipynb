```python
import json
from collections import Counter
```

```python
data = []
with open('/home/liuhao/workspace/1_googlenews/outputs/article6.json','r') as f:
    data = json.load(f)
```

```python
mlist = []
for item in data:
    url = item['url']
    baseurl = '/'.join(url.split('/')[2:3]).replace('.','_')
    mlist.append(baseurl)
common_web = [u[0] for u in Counter(mlist).most_common()[:15]]
```

```python
mlist = []
for item in data:
    url = item['url']
    baseurl = '/'.join(url.split('/')[2:3]).replace('.','_')
    if baseurl in common_web:
        print(url)
```
